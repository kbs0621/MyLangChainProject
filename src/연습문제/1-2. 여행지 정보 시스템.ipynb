```python
import os
from dotenv import load_dotenv
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI 

#load_dotenv(dotenv_path='.env')
load_dotenv()
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
print(OPENAI_API_KEY[:4])
```

```python
from langchain_core.prompts import PromptTemplate
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser
from pprint import pprint

template_text = "{model_name} 모델의 학습 원리를 {count} 문장으로 한국어로 답변해 주세요."

# PromptTemplate 인스턴스를 생성
prompt_template = PromptTemplate.from_template(template_text)

# llm = ChatOpenAI(model="gpt-3.5-turbo-0125")
llm = ChatOpenAI(
    api_key=OPENAI_API_KEY,
    base_url="https://api.groq.com/openai/v1",  # Groq API 엔드포인트
    #model="meta-llama/llama-4-scout-17b-16e-instruct",  # Spring AI와 동일한 모델
    model="openai/gpt-oss-120b",
    temperature=0.7
)
```

```python
# ----------------------
# 1단계 체인: 여행지 입력 → 대표 명소 추천
# ----------------------
step1_prompt = ChatPromptTemplate.from_messages([
    ("system", "너는 세계 여행 가이드야. 사용자가 입력한 도시나 국가에서 가장 유명한 대표 관광 명소 1곳만 추천해."),
    ("user", "여행지: {location}")
])

step1_chain = step1_prompt | llm

# ----------------------
# 2단계 체인: 명소 상세 정보 제공
# ----------------------
step2_prompt = ChatPromptTemplate.from_messages([
    ("system", "너는 역사와 문화에 정통한 여행 가이드야. 명소 이름이 주어지면 역사, 특징, 방문 팁을 구체적으로 설명해."),
    ("user", "명소: {place}")
])

step2_chain = step2_prompt | llm

# ----------------------
# Multi Chain 연결 (LCEL 방식)
# ----------------------
def multi_chain(location):
    # 1단계 실행
    step1_result = step1_chain.invoke({"location": location})
    place = step1_result.content.strip()

    # 2단계 실행
    step2_result = step2_chain.invoke({"place": place})

    # 출력 확인
    print("=== 입력 ===")
    print(location)
    print("\n=== 1단계 결과: 추천 명소 ===")
    print(place)
    print("\n=== 2단계 결과: 명소 상세 정보 ===")
    print(step2_result.content)

# ----------------------
# 실행 예시
# ----------------------
location = input("도시 또는 지역을 입력해주세요 : ")
multi_chain(location)
```
